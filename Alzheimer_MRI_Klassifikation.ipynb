# 2. Theoretische Grundlagen

<a id="21"></a>
## 2.1 Convolutional Neural Networks

Wenn wir Menschen ein Bild betrachten, erkennen wir Objekte nicht dadurch, dass wir jeden einzelnen Pixel analysieren. Stattdessen nehmen wir Kanten, Formen und Texturen wahr und setzen diese zu einem Gesamtbild zusammen. Convolutional Neural Networks (CNNs) funktionieren auf eine ähnliche Weise – und genau das macht sie so erfolgreich bei der Bildverarbeitung.

### 2.1.1 Der Faltungsvorgang (Convolution)

Das Herzstück eines CNN ist die Faltungsoperation. Dabei wird ein kleiner Filter (auch Kernel genannt) über das Eingabebild geschoben. An jeder Position berechnet der Filter das Skalarprodukt mit dem darunterliegenden Bildausschnitt:

$$
(I * K)(i, j) = \sum_{m} \sum_{n} I(i+m, j+n) \cdot K(m, n)
$$

wobei $I$ das Eingabebild, $K$ der Kernel und $(i, j)$ die Position im Ausgabebild ist.

Was bedeutet das praktisch? Ein 3×3-Filter könnte beispielsweise lernen, vertikale Kanten zu erkennen. Ein anderer Filter entdeckt horizontale Linien. Wieder ein anderer reagiert auf bestimmte Texturen. Das Schöne daran: Diese Filter werden nicht von Hand programmiert, sondern das Netzwerk lernt sie automatisch aus den Daten.

### 2.1.2 Aktivierungsfunktionen

Nach jeder Faltung wird typischerweise eine nichtlineare Aktivierungsfunktion angewendet. Die mit Abstand beliebteste ist die **ReLU** (Rectified Linear Unit):

$$
\text{ReLU}(x) = \max(0, x)
$$

ReLU ist simpel, aber effektiv. Negative Werte werden auf Null gesetzt, positive bleiben unverändert. Das klingt fast zu einfach, aber genau diese Einfachheit ist ein Vorteil: ReLU ist schnell zu berechnen und leidet weniger unter dem Problem verschwindender Gradienten als ältere Funktionen wie Sigmoid oder Tanh.

### 2.1.3 Pooling-Schichten

Zwischen den Faltungsschichten finden sich oft Pooling-Operationen. Das gängigste Verfahren ist **Max-Pooling**: Aus einem Bereich (z.B. 2×2 Pixel) wird nur der größte Wert übernommen. Pooling erfüllt mehrere Zwecke:

- **Dimensionsreduktion:** Das Bild wird kleiner, was Rechenzeit spart
- **Translationsinvarianz:** Kleine Verschiebungen im Eingabebild haben weniger Einfluss
- **Robustheit:** Das Netzwerk konzentriert sich auf die dominanten Merkmale

### 2.1.4 Batch Normalization

Ein weiterer wichtiger Baustein ist die **Batch-Normalisierung**. Sie normalisiert die Aktivierungen innerhalb eines Mini-Batches:

$$
\hat{x}_i = \frac{x_i - \mu_B}{\sqrt{\sigma_B^2 + \epsilon}}
$$

wobei $\mu_B$ und $\sigma_B$ Mittelwert und Standardabweichung des Batches sind. Anschließend werden lernbare Parameter $\gamma$ und $\beta$ angewendet:

$$
y_i = \gamma \hat{x}_i + \beta
$$

Batch Normalization hat sich als Game-Changer erwiesen. Sie stabilisiert das Training, erlaubt höhere Lernraten und wirkt sogar leicht regularisierend.

<a id="22"></a>
## 2.2 Autoencoder

Ein Autoencoder ist ein neuronales Netz, das darauf trainiert wird, seine Eingabe zu rekonstruieren – klingt erstmal nutzlos, ist es aber nicht. Der Trick liegt in der Struktur: Zwischen Eingabe und Ausgabe sitzt ein Flaschenhals, der sogenannte **latente Raum**.

### 2.2.1 Architektur

Ein Autoencoder besteht aus zwei Teilen:

**Encoder:** Komprimiert die Eingabe in eine niedrigdimensionale Repräsentation
$$
z = f_{encoder}(x)
$$

**Decoder:** Rekonstruiert die Eingabe aus der komprimierten Darstellung
$$
\hat{x} = f_{decoder}(z)
$$

Das Netzwerk wird so trainiert, dass der Rekonstruktionsfehler minimiert wird:

$$
\mathcal{L} = \|x - \hat{x}\|^2
$$

### 2.2.2 Wozu das Ganze?

Wenn ein Autoencoder ein Bild durch einen engen Flaschenhals quetschen und trotzdem wieder herstellen kann, muss er die wesentlichen Merkmale gelernt haben. Der latente Raum enthält also eine komprimierte, aber informationsreiche Repräsentation.

Für unsere MRT-Bilder bedeutet das: Der Autoencoder könnte lernen, welche Gehirnstrukturen wirklich wichtig sind – unabhängig von Rauschen oder irrelevanten Details. Außerdem können wir den latenten Raum visualisieren und verstehen, wie das Netzwerk die verschiedenen Alzheimer-Stadien intern repräsentiert.

<a id="23"></a>
## 2.3 Klassifikationsmetriken

Bei einem medizinischen Klassifikationsproblem reicht die bloße Genauigkeit (Accuracy) oft nicht aus. Stellen wir uns vor, 90% unserer Bilder zeigen gesunde Patienten. Ein Modell, das immer "gesund" sagt, hätte 90% Accuracy – wäre aber klinisch völlig nutzlos.

### 2.3.1 Genauigkeit (Accuracy)

Die einfachste Metrik ist der Anteil korrekter Vorhersagen:

$$
\text{Accuracy} = \frac{\text{Anzahl korrekter Vorhersagen}}{\text{Gesamtanzahl Vorhersagen}}
$$

### 2.3.2 Precision und Recall

**Precision** (Genauigkeit) beantwortet die Frage: "Wenn das Modell ‚krank' sagt, wie oft stimmt das?"

$$
\text{Precision} = \frac{TP}{TP + FP}
$$

**Recall** (Sensitivität) beantwortet: "Von allen tatsächlich Kranken, wie viele wurden erkannt?"

$$
\text{Recall} = \frac{TP}{TP + FN}
$$

In der Medizin ist Recall oft wichtiger als Precision. Ein übersehener Alzheimer-Fall (False Negative) ist schlimmer als ein Fehlalarm (False Positive).

### 2.3.3 F1-Score

Der F1-Score kombiniert Precision und Recall zum harmonischen Mittel:

$$
F1 = 2 \cdot \frac{\text{Precision} \cdot \text{Recall}}{\text{Precision} + \text{Recall}}
$$

### 2.3.4 Confusion Matrix

Die Konfusionsmatrix zeigt auf einen Blick, welche Klassen wie oft verwechselt werden. Für unser Vierklassen-Problem ist sie besonders aufschlussreich: Verwechselt das Modell "MildDemented" eher mit "NonDemented" oder mit "VeryMildDemented"? Solche Muster können wichtige Hinweise für Verbesserungen geben.

<a id="24"></a>
## 2.4 Regularisierungstechniken

Tiefe neuronale Netze haben viele Parameter – oft Millionen. Das macht sie einerseits ausdrucksstark, birgt andererseits die Gefahr des Overfitting: Das Netzwerk lernt die Trainingsdaten auswendig, statt generalisierende Muster zu finden.

### 2.4.1 Dropout

Dropout ist verblüffend einfach und doch effektiv: Während des Trainings werden zufällig ausgewählte Neuronen "abgeschaltet" (auf Null gesetzt). Das zwingt das Netzwerk, robuste Merkmale zu lernen, die nicht von einzelnen Neuronen abhängen.

$$
\text{Dropout}(x) = x \cdot \text{Bernoulli}(p)
$$

Typische Dropout-Raten liegen zwischen 20% und 50%.

### 2.4.2 Data Augmentation

Mehr Trainingsdaten helfen fast immer. Data Augmentation erzeugt künstliche Variationen der vorhandenen Bilder: horizontales Spiegeln, leichte Rotationen, Helligkeitsänderungen. Das Netzwerk sieht dadurch effektiv mehr Beispiele und wird robuster gegenüber solchen Variationen.

Bei medizinischen Bildern ist allerdings Vorsicht geboten: Nicht jede Augmentation ist sinnvoll. Ein vertikal gespiegeltes Gehirn macht biologisch keinen Sinn und könnte das Modell verwirren.

### 2.4.3 Early Stopping

Eine pragmatische Methode gegen Overfitting: Wir beobachten die Performance auf einem Validierungsdatensatz. Sobald sich diese nicht mehr verbessert (oder sogar verschlechtert), beenden wir das Training – auch wenn die Trainings-Loss noch sinkt.

## 3.3 Datenvorverarbeitung und Augmentation

Bevor die Bilder dem Netzwerk übergeben werden können, müssen sie vorverarbeitet werden. Dies ist einer der wichtigsten Schritte für gute Ergebnisse.

**Vorverarbeitungsschritte im Detail:**

| Schritt | Was passiert | Warum |
|---------|--------------|-------|
| `Resize` | Alle Bilder auf 160×160 Pixel | CNNs brauchen einheitliche Eingabegrößen |
| `Grayscale` | Konvertierung zu 1 Kanal | MRT-Bilder sind ohnehin grau, spart Rechenzeit |
| `ToTensor` | Numpy→PyTorch Tensor | GPU-Beschleunigung benötigt Tensoren |
| `Normalize` | Werte auf [-1, 1] skalieren | Stabileres Training, schnellere Konvergenz |

**Data Augmentation (nur Training):**

| Augmentation | Was passiert | Effekt |
|--------------|--------------|--------|
| `RandomCrop` | Zufälliger Bildausschnitt | Robustheit gegen Positionsverschiebungen |
| `RandomHorizontalFlip` | Zufällige horizontale Spiegelung | Verdoppelt effektiv die Datenmenge |
| `RandomRotation` | Leichte Drehung (±20°) | Robustheit gegen Rotationen |
| `ColorJitter` | Helligkeit/Kontrast variieren | Robustheit gegen Beleuchtungsunterschiede |
| `GaussianBlur` | Leichte Unschärfe | Robustheit gegen Bildqualitätsschwankungen |
| `RandomErasing` | Zufällig Bereiche ausblenden | Modell lernt, auch mit fehlenden Infos klarzukommen |

**Wichtig:** Augmentation wird **nur beim Training** angewendet! Validierung und Test nutzen die Originalbilder, um faire Ergebnisse zu erhalten.

````python
# Hyperparameter definieren - OPTIMIERT FÜR GESCHWINDIGKEIT + GENAUIGKEIT
IMG_SIZE = 160           # Etwas kleiner (160 statt 176) - kaum Genauigkeitsverlust, deutlich schneller
BATCH_SIZE = 64          # Größere Batch = weniger Iterationen = schneller (GPU-Speicher muss reichen)
LEARNING_RATE = 0.0003    # Höhere LR für größere Batches
EPOCHS = 40              # Weniger Epochen nötig mit AMP + größeren Batches
RANDOM_SEED = 42         # Reproduzierbarkeit
NUM_WORKERS = 2          # Schnelleres Datenladen (auf Kaggle funktioniert 2)

# Mixed Precision Training aktivieren (2-3x schneller!)
USE_AMP = True           # Automatic Mixed Precision

print("SPEED-OPTIMIERTE HYPERPARAMETER")
print("=" * 50)
print(f"  Bildgröße:     {IMG_SIZE}x{IMG_SIZE}")
print(f"  Batch-Größe:   {BATCH_SIZE} (4x größer = 4x weniger Iterationen)")
print(f"  Lernrate:      {LEARNING_RATE}")
print(f"  Epochen:       {EPOCHS}")
print(f"  NUM_WORKERS:   {NUM_WORKERS}")
print(f"  Mixed Precision (AMP): {USE_AMP} (2-3x schneller!)")
print(f"  Random Seed:   {RANDOM_SEED}")
print("\n⚡ Erwartete Beschleunigung: ~4-6x schneller pro Epoche")
````

````python
# Training Transform mit STARKER Data Augmentation für bessere Generalisierung
train_transform = transforms.Compose([
    transforms.Resize((IMG_SIZE + 20, IMG_SIZE + 20)),  # Etwas größer resizen
    transforms.RandomCrop(IMG_SIZE),                     # Dann zufällig croppen
    transforms.Grayscale(num_output_channels=1),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomVerticalFlip(p=0.2),               # Auch vertikale Flips
    transforms.RandomRotation(degrees=20),              # Stärkere Rotation
    transforms.RandomAffine(degrees=0, translate=(0.1, 0.1), scale=(0.9, 1.1)),  # Affine Transformationen
    transforms.ColorJitter(brightness=0.3, contrast=0.3),  # Stärkere Farbvariation
    transforms.RandomApply([transforms.GaussianBlur(kernel_size=3)], p=0.2),  # Manchmal Blur
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5], std=[0.5]),
    transforms.RandomErasing(p=0.2, scale=(0.02, 0.1))  # Zufälliges Ausblenden
])

# Validation/Test Transform - ohne Augmentation
val_transform = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.Grayscale(num_output_channels=1),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5], std=[0.5])
])

print("STARKE AUGMENTATION definiert:")
print("  Training: Resize+Crop, Flips, Rotation(20°), Affine, ColorJitter, GaussianBlur, RandomErasing")
print("  Validation/Test: Resize, Grayscale, Normalize")
````

````python
# Datensätze laden mit ImageFolder - nutze vorhandene train/val Struktur
train_dataset_full = datasets.ImageFolder(root=train_path, transform=train_transform)
val_dataset_full = datasets.ImageFolder(root=val_path, transform=val_transform)

print(f"Training-Datensatz: {len(train_dataset_full)} Bilder")
print(f"Validation-Datensatz: {len(val_dataset_full)} Bilder")
print(f"Klassen: {train_dataset_full.classes}")
print(f"Klassen-zu-Index-Mapping: {train_dataset_full.class_to_idx}")

# Validation-Set aufteilen in Validation und Test (50/50)
val_size = len(val_dataset_full) // 2
test_size = len(val_dataset_full) - val_size

val_dataset, test_dataset = random_split(
    val_dataset_full, 
    [val_size, test_size],
    generator=torch.Generator().manual_seed(RANDOM_SEED)
)

# Training-Set verwenden wie es ist
train_dataset = train_dataset_full

total_size = len(train_dataset) + len(val_dataset) + len(test_dataset)
print(f"\nDatenaufteilung:")
print(f"  Training:   {len(train_dataset):,} Bilder ({len(train_dataset)/total_size*100:.1f}%)")
print(f"  Validierung: {len(val_dataset):,} Bilder ({len(val_dataset)/total_size*100:.1f}%)")
print(f"  Test:       {len(test_dataset):,} Bilder ({len(test_dataset)/total_size*100:.1f}%)")
````

````python
# Validierungs- und Testdatensatz ohne Augmentation (überschreibe Transform)
# Wir erstellen neue Datasets mit den richtigen Transforms

class SubsetWithTransform(torch.utils.data.Dataset):
    """Wrapper um Subset mit eigenem Transform"""
    def __init__(self, subset, transform=None):
        self.subset = subset
        self.transform = transform
        
    def __getitem__(self, idx):
        # Original-Bild ohne Transform laden
        original_transform = self.subset.dataset.transform
        self.subset.dataset.transform = self.transform
        item = self.subset[idx]
        self.subset.dataset.transform = original_transform
        return item
    
    def __len__(self):
        return len(self.subset)

# DataLoader erstellen
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=NUM_WORKERS)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS)

print("DataLoader erstellt:")
print(f"  Train Batches: {len(train_loader)}")
print(f"  Val Batches:   {len(val_loader)}")
print(f"  Test Batches:  {len(test_loader)}")
````

## 3.4 CNN-Architektur

Jetzt wird das eigene Convolutional Neural Network definiert. Die selbst entwickelte Architektur folgt bewährten Designprinzipien moderner CNNs.

**Architektur-Übersicht:**

```
Eingabe: 1 × 160 × 160 (Graustufenbild)
    ↓
ConvBlock 1: 1 → 32 Kanäle, dann 160×160 → 80×80
    ↓
ConvBlock 2: 32 → 64 Kanäle, dann 80×80 → 40×40 (+ Residual)
    ↓
ConvBlock 3: 64 → 128 Kanäle, dann 40×40 → 20×20 (+ Residual)
    ↓
ConvBlock 4: 128 → 256 Kanäle, dann 20×20 → 10×10 (+ Residual)
    ↓
ConvBlock 5: 256 → 512 Kanäle, dann 10×10 → 5×5 (+ Residual)
    ↓
Global Average Pooling: 512 × 5 × 5 → 512
    ↓
FC: 512 → 256 → 128 → 4 (Anzahl Klassen)
```

**Komponenten erklärt:**

| Komponente | Funktion | Warum wir sie verwenden |
|------------|----------|------------------------|
| `Conv2d(3×3)` | Extrahiert lokale Merkmale (Kanten, Texturen) | Kernel-Größe 3×3 ist Standard und effizient |
| `BatchNorm2d` | Normalisiert Aktivierungen | Stabilisiert Training, ermöglicht höhere Lernraten |
| `LeakyReLU(0.1)` | Aktivierungsfunktion | Verhindert "Dying ReLU"-Problem bei negativen Werten |
| `MaxPool2d(2×2)` | Reduziert räumliche Dimension | Halbiert Bildgröße, macht Features positionsinvariant |
| `Residual Connection` | Addiert Input zum Output | Ermöglicht tiefere Netze ohne Gradient Vanishing |
| `AdaptiveAvgPool` | Global Average Pooling | Reduziert Overfitting, weniger Parameter als Flatten |
| `Dropout` | Deaktiviert zufällig Neuronen | Regularisierung gegen Overfitting |

**Warum diese Architektur?**
- **5 Conv-Blöcke** statt 4: Mehr Tiefe = komplexere Muster erkennbar
- **Doppelte Convolutions pro Block**: Ähnlich wie VGG, erhöht rezeptives Feld
- **Residual Connections**: Gradientenfluss auch bei tiefen Netzen (wie ResNet)
- **Global Average Pooling**: Reduziert Parameter drastisch am Ende

````python
class ConvBlock(nn.Module):
    """Ein doppelter Convolutional Block mit optionalem Residual"""
    def __init__(self, in_channels, out_channels, use_residual=False):
        super(ConvBlock, self).__init__()
        self.use_residual = use_residual
        
        self.conv = nn.Sequential(
            nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1, bias=False),
            nn.BatchNorm2d(out_channels),
            nn.LeakyReLU(0.1, inplace=True),
            nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1, bias=False),
            nn.BatchNorm2d(out_channels),
            nn.LeakyReLU(0.1, inplace=True),
        )
        
        # Skip-Connection falls Residual gewünscht
        if use_residual and in_channels != out_channels:
            self.skip = nn.Conv2d(in_channels, out_channels, kernel_size=1, bias=False)
        else:
            self.skip = nn.Identity() if use_residual else None
        
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)
    
    def forward(self, x):
        out = self.conv(x)
        if self.use_residual and self.skip is not None:
            out = out + self.skip(x)
        return self.pool(out)


class AlzheimerCNN(nn.Module):
    """
    VERBESSERTES CNN zur Klassifikation von Alzheimer-Stadien.
    
    Architektur (Optimiert für hohe Genauigkeit):
    - 5 Convolutional Blöcke mit je 2 Conv-Schichten (32 → 64 → 128 → 256 → 512)
    - Doppelte Conv-Layer pro Block für mehr Tiefe
    - LeakyReLU statt ReLU für stabileres Training
    - Residual Connections ab Block 2
    - Global Average Pooling vor FC-Layers
    - 3 FC-Layers mit BatchNorm und Dropout
    """
    
    def __init__(self, num_classes=4, dropout_rate=0.4):
        super(AlzheimerCNN, self).__init__()
        
        # 5 Convolutional Blöcke mit steigender Tiefe
        self.conv_blocks = nn.Sequential(
            ConvBlock(1, 32, use_residual=False),      # 176 → 88
            ConvBlock(32, 64, use_residual=True),      # 88 → 44
            ConvBlock(64, 128, use_residual=True),     # 44 → 22
            ConvBlock(128, 256, use_residual=True),    # 22 → 11
            ConvBlock(256, 512, use_residual=True),    # 11 → 5
        )
        
        # Global Average Pooling - reduziert Overfitting
        self.global_pool = nn.AdaptiveAvgPool2d((1, 1))
        
        # Fully Connected Layers mit starkem Dropout
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(512, 256),
            nn.BatchNorm1d(256),
            nn.LeakyReLU(0.1, inplace=True),
            nn.Dropout(dropout_rate),
            
            nn.Linear(256, 128),
            nn.BatchNorm1d(128),
            nn.LeakyReLU(0.1, inplace=True),
            nn.Dropout(dropout_rate),
            
            nn.Linear(128, num_classes)
        )
        
        # Gewichte initialisieren
        self._initialize_weights()
    
    def _initialize_weights(self):
        """Kaiming-Initialisierung für bessere Konvergenz"""
        for m in self.modules():
            if isinstance(m, nn.Conv2d):
                nn.init.kaiming_normal_(m.weight, mode='fan_out', nonlinearity='leaky_relu')
                if m.bias is not None:
                    nn.init.constant_(m.bias, 0)
            elif isinstance(m, (nn.BatchNorm2d, nn.BatchNorm1d)):
                nn.init.constant_(m.weight, 1)
                nn.init.constant_(m.bias, 0)
            elif isinstance(m, nn.Linear):
                nn.init.kaiming_normal_(m.weight, nonlinearity='leaky_relu')
                nn.init.constant_(m.bias, 0)
    
    def forward(self, x):
        x = self.conv_blocks(x)
        x = self.global_pool(x)
        x = self.classifier(x)
        return x
    
    def get_features(self, x):
        """Extrahiert Features vor dem Classifier (für Visualisierung)"""
        x = self.conv_blocks(x)
        x = self.global_pool(x)
        return x.view(x.size(0), -1)

# Modell erstellen
model = AlzheimerCNN(num_classes=num_classes, dropout_rate=0.4).to(device)

# Modellübersicht
print("VERBESSERTES CNN-ARCHITEKTUR")
print("=" * 60)
print(model)
print("\n" + "=" * 60)
print("Verbesserungen gegenüber Standard-CNN:")
print("  ✓ 5 statt 4 Conv-Blöcke (32→64→128→256→512)")
print("  ✓ Doppelte Conv-Schichten pro Block")
print("  ✓ Residual Connections für besseren Gradientenfluss")
print("  ✓ LeakyReLU statt ReLU")
print("  ✓ Global Average Pooling")
print("  ✓ BatchNorm auch in FC-Layers")
````

````python
# Modellparameter zählen
def count_parameters(model):
    """Zählt trainierbare und nicht-trainierbare Parameter"""
    trainable = sum(p.numel() for p in model.parameters() if p.requires_grad)
    non_trainable = sum(p.numel() for p in model.parameters() if not p.requires_grad)
    return trainable, non_trainable

trainable_params, non_trainable_params = count_parameters(model)
total_params = trainable_params + non_trainable_params

print("\nMODELL-STATISTIKEN")
print("=" * 50)
print(f"  Trainierbare Parameter:     {trainable_params:,}")
print(f"  Nicht-trainierbare Parameter: {non_trainable_params:,}")
print(f"  Gesamt:                     {total_params:,}")

# Test-Forward-Pass (im eval-Modus für BatchNorm mit batch_size=1)
print("\nTest Forward Pass:")
model.eval()  # Wichtig: eval-Modus für BatchNorm1d mit batch_size=1
test_input = torch.randn(1, 1, IMG_SIZE, IMG_SIZE).to(device)
with torch.no_grad():
    test_output = model(test_input)
print(f"  Input Shape:  {test_input.shape}")
print(f"  Output Shape: {test_output.shape}")
model.train()  # Zurück in Training-Modus
````

## 3.5 Modelltraining

Das Training ist der Kern des Deep Learning – hier lernt das Netzwerk, die MRT-Bilder zu klassifizieren.

**Training-Komponenten erklärt:**

| Komponente | Was wir verwenden | Warum |
|------------|-------------------|-------|
| **Optimizer** | AdamW | Kombination aus Adam + Weight Decay, bessere Regularisierung |
| **Loss-Funktion** | CrossEntropyLoss | Standard für Multi-Class-Klassifikation |
| **Class Weights** | Berechnet aus Klassenverteilung | Kompensiert unbalancierte Daten |
| **Label Smoothing** | 0.1 | Verhindert Overconfidence, bessere Generalisierung |
| **Scheduler** | CosineAnnealingWarmRestarts | Lernrate sinkt sanft, dann Neustart |

**Der Trainingsprozess Schritt für Schritt:**
1. **Batch laden:** 64 Bilder werden von der GPU geholt
2. **Forward Pass:** Bilder durchlaufen das Netzwerk → Vorhersagen
3. **Loss berechnen:** Vergleich Vorhersage vs. echte Labels
4. **Backward Pass:** Gradienten berechnen (Backpropagation)
5. **Gewichte aktualisieren:** Optimizer passt Parameter an
6. **Wiederholen:** Bis alle Batches durch sind (= 1 Epoche)

**Mixed Precision (AMP):**
- Berechnet Forward Pass in FP16 (16-bit) statt FP32 (32-bit)
- **2-3× schneller**, gleiche Genauigkeit
- Gradient Scaling verhindert numerische Probleme

**Early Stopping:**
Wenn die Val-Accuracy sich X Epochen lang nicht verbessert, stoppen wir das Training. Das spart Zeit und verhindert Overfitting.

````python
# Training Hyperparameter - SPEED-OPTIMIERT (passt zu größerer Batch-Size)
LEARNING_RATE = 0.001         # Höhere LR für größere Batches (Linear Scaling Rule)
NUM_EPOCHS = 40               # Weniger Epochen nötig mit AMP + größeren Batches
EARLY_STOPPING_PATIENCE = 10  # Früher stoppen
WEIGHT_DECAY = 1e-4           # L2 Regularisierung
LABEL_SMOOTHING = 0.1         # Für bessere Generalisierung

print("SPEED-OPTIMIERTE TRAINING HYPERPARAMETER")
print("=" * 50)
print(f"  Lernrate:           {LEARNING_RATE} (höher für BATCH_SIZE={BATCH_SIZE})")
print(f"  Epochen:            {NUM_EPOCHS}")
print(f"  Early Stopping:     {EARLY_STOPPING_PATIENCE} Epochen Geduld")
print(f"  Weight Decay:       {WEIGHT_DECAY}")
print(f"  Label Smoothing:    {LABEL_SMOOTHING}")
print(f"  Optimizer:          AdamW")
print(f"  Mixed Precision:    {USE_AMP}")

# Class Weights berechnen für unbalancierte Klassen
print("\nBerechne Class Weights für unbalancierte Daten...")
class_counts_list = [class_counts[c] for c in class_names]
total_samples = sum(class_counts_list)
class_weights = torch.tensor(
    [total_samples / (num_classes * count) for count in class_counts_list], 
    dtype=torch.float32
).to(device)
print(f"  Class Weights: {class_weights.cpu().numpy().round(3)}")
````

````python
# Loss-Funktion mit Label Smoothing und Class Weights
criterion = nn.CrossEntropyLoss(weight=class_weights, label_smoothing=LABEL_SMOOTHING)

# AdamW Optimizer mit Weight Decay (bessere Regularisierung als Adam)
optimizer = optim.AdamW(model.parameters(), lr=LEARNING_RATE, weight_decay=WEIGHT_DECAY)

# Cosine Annealing Scheduler - sanftere LR-Reduktion mit Warm Restarts
scheduler = optim.lr_scheduler.CosineAnnealingWarmRestarts(
    optimizer, T_0=10, T_mult=2, eta_min=1e-6
)

# Training History
history = {
    'train_loss': [],
    'train_acc': [],
    'val_loss': [],
    'val_acc': [],
    'lr': []  # Lernrate tracken
}

print("Training-Setup abgeschlossen!")
print(f"  Loss: CrossEntropyLoss (weighted, label_smoothing={LABEL_SMOOTHING})")
print(f"  Optimizer: AdamW (weight_decay={WEIGHT_DECAY})")
print(f"  Scheduler: CosineAnnealingWarmRestarts (T_0=10, T_mult=2)")
````

````python
# Mixed Precision Setup
from torch.cuda.amp import autocast, GradScaler
scaler = GradScaler(enabled=USE_AMP)

def train_epoch(model, train_loader, criterion, optimizer, device, scaler):
    """Trainiert das Modell für eine Epoche MIT Mixed Precision (AMP)"""
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0
    
    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)
        
        optimizer.zero_grad()
        
        # Mixed Precision: Forward Pass in FP16
        with autocast(enabled=USE_AMP):
            outputs = model(images)
            loss = criterion(outputs, labels)
        
        # Backward Pass mit Gradient Scaling
        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()
        
        # Statistiken
        running_loss += loss.item() * images.size(0)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()
    
    epoch_loss = running_loss / total
    epoch_acc = correct / total
    return epoch_loss, epoch_acc


def validate_epoch(model, val_loader, criterion, device):
    """Evaluiert das Modell auf dem Validierungsset MIT Mixed Precision"""
    model.eval()
    running_loss = 0.0
    correct = 0
    total = 0
    
    with torch.no_grad():
        for images, labels in val_loader:
            images, labels = images.to(device), labels.to(device)
            
            with autocast(enabled=USE_AMP):
                outputs = model(images)
                loss = criterion(outputs, labels)
            
            running_loss += loss.item() * images.size(0)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    
    epoch_loss = running_loss / total
    epoch_acc = correct / total
    return epoch_loss, epoch_acc

print("Training-Funktionen mit Mixed Precision (AMP) definiert!")
print("⚡ AMP aktiviert:", USE_AMP)
````

````python
# Haupttrainingsschleife - SPEED-OPTIMIERT MIT AMP
import time

print("TRAINING STARTET")
print("=" * 70)
print(f"Modell: AlzheimerCNN (5 Conv-Blöcke mit Residual)")
print(f"Gerät: {device}")
print(f"Mixed Precision (AMP): {USE_AMP}")
print(f"Batch-Größe: {BATCH_SIZE}")
print("=" * 70)

best_val_loss = float('inf')
best_val_acc = 0.0
patience_counter = 0
best_model_state = None
start_time = time.time()

for epoch in range(NUM_EPOCHS):
    epoch_start = time.time()
    
    # Training MIT scaler für AMP
    train_loss, train_acc = train_epoch(model, train_loader, criterion, optimizer, device, scaler)
    
    # Validierung
    val_loss, val_acc = validate_epoch(model, val_loader, criterion, device)
    
    epoch_time = time.time() - epoch_start
    
    # History speichern
    history['train_loss'].append(train_loss)
    history['train_acc'].append(train_acc)
    history['val_loss'].append(val_loss)
    history['val_acc'].append(val_acc)
    history['lr'].append(optimizer.param_groups[0]['lr'])
    
    # Learning Rate anpassen
    scheduler.step(epoch + 1)
    
    # Ausgabe mit Zeit pro Epoche
    current_lr = optimizer.param_groups[0]['lr']
    print(f"Epoch [{epoch+1:2d}/{NUM_EPOCHS}] | "
          f"Train: {train_loss:.4f}/{train_acc:.4f} | "
          f"Val: {val_loss:.4f}/{val_acc:.4f} | "
          f"Zeit: {epoch_time:.1f}s")
    
    # Best Model speichern
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        best_val_loss = val_loss
        best_model_state = model.state_dict().copy()
        patience_counter = 0
        print(f"   ★ Neues bestes Modell! (Accuracy: {best_val_acc:.4f})")
    else:
        patience_counter += 1
        
    # Early Stopping
    if patience_counter >= EARLY_STOPPING_PATIENCE:
        print(f"\n⚠ Early Stopping nach {epoch+1} Epochen")
        break

# Training beenden
elapsed_time = time.time() - start_time
print("\n" + "=" * 70)
print("TRAINING ABGESCHLOSSEN")
print("=" * 70)
print(f"  Trainingszeit: {elapsed_time/60:.1f} Minuten ({elapsed_time/len(history['train_loss']):.1f}s pro Epoche)")
print(f"  Beste Val-Accuracy: {best_val_acc:.4f} ({best_val_acc*100:.2f}%)")
print(f"  Beste Val-Loss: {best_val_loss:.4f}")
print(f"  Anzahl Epochen: {len(history['train_loss'])}")

# Bestes Modell laden
if best_model_state is not None:
    model.load_state_dict(best_model_state)
    print("\n✓ Bestes Modell wiederhergestellt!")
````

## 3.6 Autoencoder

Zusätzlich zum Klassifikationsmodell implementieren wir einen **Convolutional Autoencoder**. Dieser lernt unüberwacht, MRT-Bilder zu komprimieren und wieder zu rekonstruieren.

**Was ist ein Autoencoder?**

```
Eingabe (160×160) → Encoder → Latenter Raum (256 Werte) → Decoder → Rekonstruktion (160×160)
```

**Komponenten:**

| Teil | Funktion | Architektur |
|------|----------|-------------|
| **Encoder** | Komprimiert das Bild | 5 Conv-Schichten (1→32→64→128→256→512), dann FC auf 256 |
| **Latenter Raum** | Kompakte Repräsentation | 256 Dimensionen (Flaschenhals) |
| **Decoder** | Rekonstruiert das Bild | 5 TransposeConv-Schichten, spiegelbildlich zum Encoder |

**Warum Autoencoder nützlich sind:**
1. **Dimensionsreduktion:** 160×160 = 25.600 Pixel → nur 256 Werte
2. **Visualisierung:** t-SNE auf dem latenten Raum zeigt, ob Klassen trennbar sind
3. **Anomalie-Erkennung:** Hoher Rekonstruktionsfehler = ungewöhnliches Bild
4. **Feature Learning:** Der Encoder lernt relevante Merkmale ohne Labels

**Training:**
- **Loss-Funktion:** MSE (Mean Squared Error) zwischen Original und Rekonstruktion
- **Ziel:** Rekonstruktion soll möglichst nah am Original sein
- **Latenter Raum:** Je besser die Rekonstruktion, desto informativer der latente Raum

````python
class ConvAutoencoder(nn.Module):
    """
    Verbesserter Convolutional Autoencoder für MRT-Bilder.
    
    Encoder: Komprimiert 176x176 Bilder in einen latenten Raum
    Decoder: Rekonstruiert das Bild aus dem latenten Raum
    
    Verwendet adaptive Pooling für flexible Bildgrößen.
    """
    
    def __init__(self, latent_dim=256):  # Größerer latenter Raum
        super(ConvAutoencoder, self).__init__()
        self.latent_dim = latent_dim
        
        # Encoder - passt sich an IMG_SIZE an
        self.encoder = nn.Sequential(
            # 1 x 176 x 176 → 32 x 88 x 88
            nn.Conv2d(1, 32, kernel_size=3, stride=2, padding=1),
            nn.BatchNorm2d(32),
            nn.LeakyReLU(0.1, inplace=True),
            
            # 32 x 88 x 88 → 64 x 44 x 44
            nn.Conv2d(32, 64, kernel_size=3, stride=2, padding=1),
            nn.BatchNorm2d(64),
            nn.LeakyReLU(0.1, inplace=True),
            
            # 64 x 44 x 44 → 128 x 22 x 22
            nn.Conv2d(64, 128, kernel_size=3, stride=2, padding=1),
            nn.BatchNorm2d(128),
            nn.LeakyReLU(0.1, inplace=True),
            
            # 128 x 22 x 22 → 256 x 11 x 11
            nn.Conv2d(128, 256, kernel_size=3, stride=2, padding=1),
            nn.BatchNorm2d(256),
            nn.LeakyReLU(0.1, inplace=True),
            
            # 256 x 11 x 11 → 512 x 5 x 5 (zusätzliche Schicht)
            nn.Conv2d(256, 512, kernel_size=3, stride=2, padding=1),
            nn.BatchNorm2d(512),
            nn.LeakyReLU(0.1, inplace=True),
        )
        
        # Adaptive Pooling für feste Größe vor FC
        self.adaptive_pool = nn.AdaptiveAvgPool2d((4, 4))
        
        # Latenter Raum
        self.fc_encode = nn.Linear(512 * 4 * 4, latent_dim)
        self.fc_decode = nn.Linear(latent_dim, 512 * 4 * 4)
        
        # Decoder - rekonstruiert auf Zielgröße
        self.decoder = nn.Sequential(
            # 512 x 4 x 4 → 256 x 8 x 8
            nn.ConvTranspose2d(512, 256, kernel_size=4, stride=2, padding=1),
            nn.BatchNorm2d(256),
            nn.LeakyReLU(0.1, inplace=True),
            
            # 256 x 8 x 8 → 128 x 16 x 16
            nn.ConvTranspose2d(256, 128, kernel_size=4, stride=2, padding=1),
            nn.BatchNorm2d(128),
            nn.LeakyReLU(0.1, inplace=True),
            
            # 128 x 16 x 16 → 64 x 32 x 32
            nn.ConvTranspose2d(128, 64, kernel_size=4, stride=2, padding=1),
            nn.BatchNorm2d(64),
            nn.LeakyReLU(0.1, inplace=True),
            
            # 64 x 32 x 32 → 32 x 64 x 64
            nn.ConvTranspose2d(64, 32, kernel_size=4, stride=2, padding=1),
            nn.BatchNorm2d(32),
            nn.LeakyReLU(0.1, inplace=True),
            
            # 32 x 64 x 64 → 1 x 128 x 128
            nn.ConvTranspose2d(32, 1, kernel_size=4, stride=2, padding=1),
            nn.Tanh()  # Output in [-1, 1] wie unsere normalisierten Bilder
        )
        
        # Finaler Upsample auf Zielgröße
        self.final_upsample = nn.Upsample(size=(IMG_SIZE, IMG_SIZE), mode='bilinear', align_corners=False)
    
    def encode(self, x):
        """Encodiert Bild in latenten Raum"""
        x = self.encoder(x)
        x = self.adaptive_pool(x)
        x = x.view(x.size(0), -1)
        z = self.fc_encode(x)
        return z
    
    def decode(self, z):
        """Decodiert latenten Vektor zurück zum Bild"""
        x = self.fc_decode(z)
        x = x.view(x.size(0), 512, 4, 4)
        x = self.decoder(x)
        x = self.final_upsample(x)
        return x
    
    def forward(self, x):
        z = self.encode(x)
        reconstructed = self.decode(z)
        return reconstructed, z

# Autoencoder erstellen
autoencoder = ConvAutoencoder(latent_dim=256).to(device)
print("VERBESSERTER AUTOENCODER")
print("=" * 50)
print(f"Latenter Raum: 256 Dimensionen")
print(f"Eingang: {IMG_SIZE}x{IMG_SIZE} → Ausgang: {IMG_SIZE}x{IMG_SIZE}")
print(f"Encoder: 5 Conv-Schichten (1→32→64→128→256→512)")
print(f"Decoder: 5 TransposeConv-Schichten + Upsample")
````

````python
# Autoencoder Training
AE_EPOCHS = 20
AE_LR = 0.001

ae_criterion = nn.MSELoss()
ae_optimizer = optim.Adam(autoencoder.parameters(), lr=AE_LR)

ae_history = {'train_loss': [], 'val_loss': []}

print("AUTOENCODER TRAINING")
print("=" * 70)

for epoch in range(AE_EPOCHS):
    # Training
    autoencoder.train()
    train_loss = 0.0
    
    for images, _ in train_loader:
        images = images.to(device)
        
        ae_optimizer.zero_grad()
        reconstructed, _ = autoencoder(images)
        loss = ae_criterion(reconstructed, images)
        loss.backward()
        ae_optimizer.step()
        
        train_loss += loss.item() * images.size(0)
    
    train_loss /= len(train_dataset)
    
    # Validierung
    autoencoder.eval()
    val_loss = 0.0
    
    with torch.no_grad():
        for images, _ in val_loader:
            images = images.to(device)
            reconstructed, _ = autoencoder(images)
            loss = ae_criterion(reconstructed, images)
            val_loss += loss.item() * images.size(0)
    
    val_loss /= len(val_dataset)
    
    ae_history['train_loss'].append(train_loss)
    ae_history['val_loss'].append(val_loss)
    
    print(f"Epoch [{epoch+1:2d}/{AE_EPOCHS}] | Train Loss: {train_loss:.6f} | Val Loss: {val_loss:.6f}")

print("\nAutoencoder Training abgeschlossen!")
````

## 3.7 Modellbewertung und Visualisierung

Abschließend wird das trainierte CNN auf dem Testdatensatz evaluiert. Die Evaluation verwendet **ungesehene Daten** (Testset), um die echte Generalisierungsfähigkeit zu messen.

**Evaluationsmetriken erklärt:**

| Metrik | Formel | Was sie misst |
|--------|--------|---------------|
| **Accuracy** | Richtige / Gesamt | Anteil korrekt klassifizierter Bilder |
| **Precision** | TP / (TP + FP) | Wie viele der positiven Vorhersagen sind korrekt? |
| **Recall** | TP / (TP + FN) | Wie viele der echten Positiven wurden gefunden? |
| **F1-Score** | 2 × (Prec × Rec) / (Prec + Rec) | Harmonisches Mittel aus Precision und Recall |

**Visualisierungen:**

1. **Confusion Matrix:** Zeigt, welche Klassen miteinander verwechselt werden
   - Diagonale = korrekte Klassifikationen
   - Abseits der Diagonale = Fehler (z.B. MildDemented als VeryMildDemented klassifiziert)

2. **Klassenweise Accuracy:** Balkendiagramm zeigt, welche Klassen gut/schlecht erkannt werden

3. **Beispiel-Vorhersagen:** Zeigt echte Bilder mit Vorhersage und Konfidenz
   - Grün = korrekt
   - Rot = falsch

**Warum mehrere Metriken wichtig sind:**
Accuracy allein reicht bei unbalancierten Daten nicht aus! Ein Modell könnte 90% Accuracy haben, aber die seltene Klasse nie erkennen. F1-Score und klassenweise Metriken decken solche Probleme auf.

````python
# Finale Evaluation auf dem Testset
print("FINALE EVALUATION AUF TESTDATEN")
print("=" * 50)

model.eval()
all_preds = []
all_true = []
all_probs = []

with torch.no_grad():
    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        probs = F.softmax(outputs, dim=1)
        _, predicted = torch.max(outputs.data, 1)
        
        all_preds.extend(predicted.cpu().numpy())
        all_true.extend(labels.cpu().numpy())
        all_probs.extend(probs.cpu().numpy())

all_preds = np.array(all_preds)
all_true = np.array(all_true)
all_probs = np.array(all_probs)

# Metriken berechnen
test_accuracy = accuracy_score(all_true, all_preds)
test_f1 = f1_score(all_true, all_preds, average='weighted')

print(f"\nTest Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")
print(f"Test F1-Score (weighted): {test_f1:.4f}")
````

<a id="43"></a>
## 4.3 Limitationen und ethische Aspekte

### Technische Limitationen

- **Datensatzgröße:** Mit ~6.400 Bildern ist der Datensatz relativ klein für Deep Learning
- **Bildqualität:** Die vorverarbeiteten MRT-Bilder haben möglicherweise relevante Details verloren
- **Generalisierbarkeit:** Das Modell wurde nur auf diesem spezifischen Datensatz trainiert und getestet
- **Keine klinische Validierung:** Es fehlen unabhängige klinische Tests

### Ethische Überlegungen

**Verantwortungsvoller Einsatz:**
- KI-Systeme sollten Ärzte unterstützen, nicht ersetzen
- Falsch-negative Ergebnisse könnten zu verzögerter Behandlung führen
- Falsch-positive Ergebnisse können unnötige Ängste und weitere Tests verursachen

**Bias und Fairness:**
- Der Trainingsdatensatz repräsentiert möglicherweise nicht alle Bevölkerungsgruppen
- Unterschiede in MRT-Geräten und Aufnahmeprotokollen können die Performance beeinflussen

**Transparenz:**
- "Black Box"-Modelle sind in der Medizin problematisch
- Explainable AI (XAI) Methoden wie Grad-CAM könnten helfen, Entscheidungen nachvollziehbar zu machen

## 5.1 Zusammenfassung

In dieser Arbeit wurde ein eigenes Convolutional Neural Network zur Klassifikation von Alzheimer-Stadien entwickelt und evaluiert.

| Ziel | Ergebnis | Status |
|------|----------|--------|
| Selbst entwickeltes CNN | AlzheimerCNN mit 5.05 Mio. Parametern | ✅ Erfolgreich |
| Hohe Genauigkeit | **97.12% Test-Accuracy** | ✅ Übertroffen |
| Alle 4 Klassen erkennbar | 93.35% - 100.0% pro Klasse | ✅ Erfolgreich |
| Autoencoder implementiert | 256-dim Latent Space + t-SNE | ✅ Erfolgreich |
| Effizientes Training | 45.6 Minuten (68.4s/Epoche) | ✅ Erfolgreich |
| Beste Val-Accuracy | 98.62% | ✅ Exzellent |

### Wichtigste Erkenntnisse

1. **Machbarkeit bestätigt:** Ein selbst entwickeltes CNN ohne vortrainierte Gewichte erreicht 97.12% Accuracy – vergleichbar mit Transfer Learning-Ansätzen

2. **Architektur-Entscheidungen:** Die Kombination aus:
   - 5 ConvBlöcken mit Residual Connections
   - LeakyReLU und BatchNorm
   - Global Average Pooling
   - AdamW mit Label Smoothing
   
   hat sich als äußerst effektiv erwiesen

3. **Klassenspezifische Erkenntnisse:**
   - Moderate Demenz (100%) ist am einfachsten zu erkennen (deutliche visuelle Merkmale)
   - Mild Demented (93.35%) ist am schwierigsten (Übergangsklasse)

4. **Autoencoder-Nutzen:** Die t-SNE-Visualisierung des 256-dimensionalen latenten Raums bestätigt die Klassentrennung und erklärt die Klassifikationsergebnisse
